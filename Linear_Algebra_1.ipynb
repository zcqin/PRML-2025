```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create a unit circle
theta = np.linspace(0, 2 * np.pi, 100)
circle = np.array([np.cos(theta), np.sin(theta)])

# Step 2: Define the transformation matrix A
A = np.array([[2, 3],
              [1, 1]])

# Step 3: Calculate eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

# Step 4: Apply the linear transformation
ellipse = A @ circle

# Step 5: Plot the original circle and the transformed ellipse
fig, ax = plt.subplots(1, 2, figsize=(10, 5), subplot_kw={'aspect': 'equal'})

# Original circle with eigenvectors
ax[0].plot(circle[0, :], circle[1, :], 'b')
ax[0].quiver([0, 0], [0, 0], eigenvectors[0, :], eigenvectors[1, :], angles='xy', scale_units='xy', scale=1, color=['black'])
ax[0].set_xlim([-3, 3])
ax[0].set_ylim([-3, 3])
ax[0].grid(True)
ax[0].text(eigenvectors[0, 0], eigenvectors[1, 0], r'$v^{(1)}$', fontsize=14, color='black')
ax[0].text(eigenvectors[0, 1], eigenvectors[1, 1], r'$v^{(2)}$', fontsize=14, color='black')
ax[0].set_title('Before multiplication')

# Transformed ellipse with scaled eigenvectors
ax[1].plot(ellipse[0, :], ellipse[1, :], 'b')
for i in range(len(eigenvalues)):
    ax[1].quiver([0, 0], [0, 0], eigenvalues[i] * eigenvectors[0, i], eigenvalues[i] * eigenvectors[1, i], angles='xy', scale_units='xy', scale=1, color=['black'])

ax[1].set_xlim([-4, 4])
ax[1].set_ylim([-4, 4])
ax[1].grid(True)
ax[1].text(eigenvectors[0, 0], eigenvalues[0] * eigenvectors[1, 0], r'$\lambda v^{(1)}$', fontsize=14, color='black')
ax[1].text(eigenvectors[0, 1], eigenvalues[1] * eigenvectors[1, 1], r'$\lambda v^{(2)}$', fontsize=14, color='black')
ax[1].set_title('After multiplication')

# Show plot with equal aspect ratio
plt.show()
```

```python
plt.style.use('seaborn-poster')

%matplotlib inline

def plot_vect(x, b, xlim, ylim):
    '''
    function to plot two vectors,
    x - the original vector
    b - the transformed vector
    xlim - the limit for x
    ylim - the limit for y
    '''
    plt.figure(figsize = (10, 6))
    plt.quiver(0,0,x[0],x[1],\
        color='k',angles='xy',\
        scale_units='xy',scale=1,\
        label='Original vector')
    plt.quiver(0,0,b[0],b[1],\
        color='g',angles='xy',\
        scale_units='xy',scale=1,\
        label ='Transformed vector')
    plt.xlim(xlim)
    plt.ylim(ylim)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
    plt.show()
```

```python

# Lp norm function
def lp_norm(x, y, p):
    return (np.abs(x)**p + np.abs(y)**p)**(1/p)

# Generate a grid of points
x = np.linspace(-1.2, 1.2, 400)
y = np.linspace(-1.2, 1.2, 400)
X, Y = np.meshgrid(x, y)

# Values of p for which to plot the contours
p_values = [1, 2, 3, 5, np.inf] # Including L1, L2, and L∞ norms
print(p_values)
color_values = ['black', 'blue', 'green', 'red', 'purple', 'orange']
print(color_values)

# Plot the contours
plt.figure(figsize=(6, 6))
index = 0


for p in p_values:
    if p == np.inf:
        Z = np.maximum(np.abs(X), np.abs(Y))
    else:
        Z = lp_norm(X, Y, p)
    con = plt.contour(X, Y, Z, levels=[1], colors=color_values[index])
 #   plt.clabel(con, inline=True, fontsize=8)
    index = index + 1

plt.title('Contours of Constant L-p Norms')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.axis('equal')
plt.grid(True)
plt.legend()
plt.show()

```

Linear Transformation in 2D space by multiplying a matrix

```python


# Define the original vector
v = np.array([2, 2])

# Define a linear transformation matrix
A = np.array([[-4, -2],
              [2, 1]])

# Apply the linear transformation
v_transformed = A.dot(v)

# Define a nonlinear transformation function
def nonlinear_transformation(x):
    return np.array([x[0]**2 - x[1]**2, 2*x[0]*x[1]])

# Apply the nonlinear transformation
v_nonlinear_transformed = nonlinear_transformation(v)

# Plotting
plt.figure(figsize=(8, 6))

# Original vector
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Original Vector')

# Linearly transformed vector
plt.quiver(0, 0, v_transformed[0], v_transformed[1], angles='xy', scale_units='xy', scale=1, color='green', label='Linearly Transformed')

# Non-linearly transformed vector
plt.quiver(0, 0, v_nonlinear_transformed[0], v_nonlinear_transformed[1], angles='xy', scale_units='xy', scale=1, color='red', label='Non-linearly Transformed')

plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.title('Vector Transformation in 2D Space')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
