13 Object-Oriented Programming


```python
class Person():
    def __init__(self, name, age, height, weight):
        self.name = name
        self.age = age
        self.height = height
        self.weight = weight


    def greet(self):
        print("Greetings, " + self.name)
```

```python
person1 = Person(name = 'Iron Man', age = 35, height = 175, weight = 75)
person2 = Person(name = 'Wonder Woman', age = 25, height = 170, weight = 55)
person3 = Person(name = 'American Captain', age = 32, height = 180, weight = 80)
person4 = Person(name = 'Superman', age = 75, height = 185, weight = 78)


person1.greet()
person2.greet()

print(person3.name)
print(person4.age)
```

14. Computational Complexity


```python
import numpy as np
import matplotlib.pyplot as plt


#numbers = range(1, 100)
numbers = np.arange(1, 1e2)

#squared_numbers = [x ** 2 for x in numbers]
squared_numbers = numbers ** 2

plt.plot(numbers, squared_numbers)
plt.yscale('log')
plt.show()


plt.figure(figsize = (10, 8))
n = np.arange(1, 1e3)

plt.plot(np.log(n), label = 'log(n)')
plt.plot(n, label = 'n')
plt.plot(n**2, label = '$n^2$')
# plt.plot(2**n, label = '$2^n$')
plt.yscale('log')

plt.legend()
plt.show()
```

```python
# Multi-subfigure plot example

x = np.arange(11)
y = x**2

plt.figure(figsize = (14, 8))

plt.subplot(2, 3, 1)
plt.plot(x,y)
plt.title('Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()

plt.subplot(2, 3, 2)
plt.scatter(x,y)
plt.title('Scatter')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()

plt.subplot(2, 3, 3)
plt.bar(x,y)
plt.title('Bar')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()

plt.subplot(2, 3, 4)
plt.loglog(x,y)
plt.title('Loglog')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(which='both')

plt.subplot(2, 3, 5)
plt.semilogx(x,y)
plt.title('Semilogx')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(which='both')

plt.subplot(2, 3, 6)
plt.semilogy(x,y)
plt.title('Semilogy')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()

plt.tight_layout()

plt.show()
```

16. 3D Plotting

```python
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
#plt.style.use('seaborn-poster')


fig = plt.figure(figsize = (6,6))
ax = plt.axes(projection='3d')
ax.grid()

t = np.arange(0, 10*np.pi, np.pi/50)
x = np.sin(t)
y = np.cos(t)

ax.plot3D(x, y, t)
ax.set_title('3D Parametric Plot')

# Set axes label
ax.set_xlabel('x = sin(t)', labelpad=20)
ax.set_ylabel('y = cons(t)' , labelpad=20)
ax.set_zlabel('t', labelpad=20)

plt.show()
```

```python
fig = plt.figure(figsize = (12,10))
ax = plt.axes(projection='3d')

x = np.arange(-5, 5.1, 0.2)
y = np.arange(-5, 5.1, 0.2)

X, Y = np.meshgrid(x, y)
Z = np.sin(X)*np.cos(Y)

surf = ax.plot_surface(X, Y, Z, cmap = plt.cm.cividis)

# Set axes label
ax.set_xlabel('x', labelpad=20)
ax.set_ylabel('y', labelpad=20)
ax.set_zlabel('z', labelpad=20)

fig.colorbar(surf, shrink=0.5, aspect=8)

plt.show()
```
